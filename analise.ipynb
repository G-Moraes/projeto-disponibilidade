```python
import pandas as pd
import datetime as dt
```

# 1 - ler escala

```python
escala = pd.read_csv('escala.csv')

#escala.sort_values(by = ['shift_date', 'member_id'], inplace = True)

escala
```

## 1.1 - tratar dados

```python
start_time_interval = []
end_time_interval = []

for text in escala['time_interval']:
    
    start = text.index('X')
    end1 = text.index('m_', start)
    end2 = text.index('m__', end1)

    startInterval = text[start+1:end1]
    endInterval = text[end1+2: end2]

    startInterval = startInterval.replace('h', ':')
    endInterval = endInterval.replace('h', ':')

    hora_start = dt.datetime.strptime(startInterval, "%H:%M").time()
    hora_end = dt.datetime.strptime(endInterval, "%H:%M").time()

    '''hora_start = dt.timedelta(hours = hora_start.hour, minutes = hora_start.minute)

    hora_end = dt.timedelta(hours = hora_end.hour, minutes = hora_end.minute)'''

    start_time_interval.append(hora_start)
    end_time_interval.append(hora_end)

escala.insert(3, 'start_time_interval', start_time_interval)
escala.insert(4, 'end_time_interval', end_time_interval)
```

```python
escala.drop(columns = 'time_interval', inplace=True)
```

```python
escala.sort_values(by = ['shift_date', 'member_id', 'start_time_interval', 'end_time_interval'], inplace = True)

escala
```

# 2 - membros

```python
membros = pd.read_csv('membros.csv')

for index in membros['Data'].index:

    data_dt = dt.datetime.strptime(membros.loc[index]['Data'], "%Y-%m-%d").date()

    membros.loc[index]['Data'] = data_dt

membros.sort_values(by = ['Data', 'member_id'], inplace = True)

membros
```

# 3 - relatorios

```python
relatorio = pd.read_csv('relatorio_log_in_e_out.csv')

relatorio.sort_values(by = ['created_at', 'member_id'], inplace = True)

relatorio
```

## 3.1 - tratar dados

```python
dates = []

for date in  relatorio['created_at']:

    aux_date = date.split()[0]

    dates.append(dt.datetime.strptime(aux_date, '%Y-%m-%d').date())

times = []

for time in relatorio['created_at']:

    time_aux_space = time.split()[1]
    time_aux_dot = time_aux_space.split('.', 1)[0]
    time_aux_plus = time_aux_dot.split('+', 1)[0]
    times.append(dt.datetime.strptime(time_aux_plus, "%H:%M:%S").time())

```

```python
relatorio.insert(2, 'created_date', dates)
relatorio.insert(3, 'created_time', times)
```

```python
relatorio.drop(columns = 'created_at', inplace = True)
```

```python
relatorio.sort_values(by = ['created_date', 'member_id'], inplace = True)

relatorio
```

# 4 - informaÃ§Ã£o A

```python
#celula gera a informaÃ§Ã£o A

#1 - salvo um dataframe com cada membro que estava escalado para atendimento e que dias eles estÃ£o
#2 - checo membro um por um, e quantos minutos ele estava escalado para atender em dias especifico
    #2.1 - protip: escalamento Ã© equivalente a meia hora, entÃ£o sÃ³ (entrada x 30) para obter minutos
#3 - armazeno essa informaÃ§Ã£o em um dataframe

attending_member = escala.loc[escala['resulting_status'] == 'A']

#unique_attending_members = attending_member['member_id'].unique()

df_atendimento = pd.DataFrame(columns = ['member_id', 'Data', 'escala_atendimento'])

for member in attending_member['member_id'].unique():

    df_attending_member = attending_member.loc[attending_member['member_id'] == member]
    
    for date in df_attending_member['shift_date'].unique():

        concatenado = pd.DataFrame(columns = ['member_id', 'Data', 'escala_atendimento'])

        aux = df_attending_member.loc[df_attending_member['shift_date'] == date]

        concatenado['member_id'] = aux['member_id']
        concatenado['Data'] = aux['shift_date']
        concatenado['escala_atendimento'] = len(aux) * 30

        concatenado.drop_duplicates(inplace = True)

        df_atendimento = pd.concat([df_atendimento, concatenado])
```

# informaÃ§Ã£o C

```python
#etapa 1, pego todos os membros que estÃ£o como ativos
membros_ativos = membros.loc[membros['Status'] == 'Ativo']

#etapa 3
unique_members = relatorio['member_id'].unique()

#crio a escalaÃ§Ã£o em minutos por dia de cada encantador
df_disponibilidade = pd.DataFrame(columns = ['member_id', 'Data', 'Escala'])

for member in membros_ativos['member_id'].unique():

    concatenado = pd.DataFrame(columns = ['member_id', 'Data'])

    aux_membros_ativos = membros_ativos.loc[membros_ativos['member_id'] == member]

    aux_escala = escala.loc[escala['member_id'] == member]

    aux_escala.reset_index(inplace = True)

    lista_disp = []

    for data in aux_escala['shift_date'].unique():

        aux_escala_data = aux_escala.loc[aux_escala['shift_date'] == data]

        lista_disp.append(len(aux_escala_data) * 30)  

    concatenado['Data'] = aux_escala['shift_date'].unique()

    concatenado['Escala'] = lista_disp

    lista_member = []

    for index in range(len(aux_escala['shift_date'].unique())):

        lista_member.append(aux_escala.loc[index]['member_id'])

    concatenado['member_id'] = lista_member    

    concatenado.drop_duplicates()

    df_disponibilidade = pd.concat([df_disponibilidade, concatenado])
```

```python
#calculo a disponibilidade por dia de cada encantador
lista_disp = []

for member in df_disponibilidade['member_id'].unique():

    df_aux_disponibilidade = df_disponibilidade.loc[df_disponibilidade['member_id'] == member]

    df_relatorio = relatorio.loc[relatorio['member_id'] == member]

    df_relatorio.reset_index(inplace = True)

    for data in df_aux_disponibilidade['Data']:

        dt_data = dt.datetime.strptime(data, '%Y-%m-%d').date()

        df_aux_relatorio = df_relatorio.loc[df_relatorio['created_date'] == dt_data]

        df_aux_relatorio.reset_index(inplace = True)

        i = 0

        tempo_disponivel_dia = dt.timedelta(0)

        while i < len(df_aux_disponibilidade['Data']):

            #procuro os momentos em que ele estava disponÃ­vel
            try:

                #checo a primeira ocorrÃªncia de DisponÃ­vel no relatÃ³rio do membro naquele dia

                index_disponivel = list(df_aux_relatorio['agent_status']).index('DisponÃ­vel', i, len(df_aux_relatorio))

                if(index_disponivel == -1 or index_disponivel + 1 >= len(df_aux_relatorio) or i + 1 >= len(df_aux_relatorio)):
                    break
                
                #se o prÃ³ximo status nÃ£o ainda estiver no escopo 
                elif(i + 1 <= len(df_aux_relatorio) and index_disponivel + 1 <= len(df_aux_relatorio)):

                    i = index_disponivel + 1

                    prox_tempo = df_aux_relatorio.loc[i]['created_time']

                    prox_tempo = dt.timedelta(hours = prox_tempo.hour, minutes = prox_tempo.minute, seconds = prox_tempo.second)

                    
                    tempo_disponivel = df_aux_relatorio.loc[index_disponivel]['created_time']
    
                    tempo_disponivel = dt.timedelta(hours = tempo_disponivel.hour, minutes = tempo_disponivel.minute, seconds = tempo_disponivel.second)

                    tempo_disponivel_dia += prox_tempo - tempo_disponivel
   
            except ValueError:
                
                break
        
        aux_tempo = str(tempo_disponivel_dia).split('0 days ')

        if(len(aux_tempo) > 1):
            aux_string = str(tempo_disponivel_dia).split('0 days ')[1]
            aux_dt = dt.datetime.strptime(aux_string, "%H:%M:%S")

        else:
            aux_dt = dt.datetime.strptime(str(tempo_disponivel_dia), "%H:%M:%S")

        tempo_disponivel_dia = dt.timedelta(hours = aux_dt.hour, minutes = aux_dt.minute, seconds = aux_dt.second)

        lista_disp.append(tempo_disponivel_dia)

lista_minutes = []

for element in lista_disp:

    lista_minutes.append(int(element.total_seconds() / 60))

df_disponibilidade['Logado por dia'] = lista_minutes
```

# InformaÃ§Ã£o G

```python
#utilizo a tabela escala, mas sÃ³ com os membros que estÃ£o como atendimento

lista_horarios = []

for member in df_disponibilidade['member_id'].unique():

    concatenado = pd.DataFrame(columns = ['member_id', 'Data', 'Primeiro HorÃ¡rio'])

    df_escala = escala.loc[escala['member_id'] == member]

    df_aux_disponibilidade = df_disponibilidade.loc[df_disponibilidade['member_id'] == member]

    for data in df_aux_disponibilidade['Data']:

        df_escala_data = df_escala.loc[df_escala['shift_date'] == data]

        primeiro_horario = list(df_escala_data['start_time_interval'])[0]

        lista_horarios.append(primeiro_horario)

df_horario = pd.DataFrame(columns = ['member_id', 'Data', 'Primeiro HorÃ¡rio'])

df_horario['member_id'] = df_disponibilidade['member_id']

df_horario['Data'] = df_disponibilidade['Data']

df_horario['Primeiro HorÃ¡rio'] = lista_horarios
```

```python
lista_atrasado = []

lista_resultado_atraso = []

for member in df_horario['member_id'].unique():

    aux_relatorio = relatorio.loc[relatorio['member_id'] == member]

    aux_df_horario = df_horario.loc[df_horario['member_id'] == member]

    for data in aux_df_horario['Data']:

        dt_data = dt.datetime.strptime(data,'%Y-%m-%d').date()

        relatorio_data_aux = aux_relatorio.loc[aux_relatorio['created_date'] == dt_data]

        relatorio_data_aux = relatorio_data_aux.sort_values(by = 'created_time')

        lista_entrada_log = list(relatorio_data_aux.loc[relatorio_data_aux['agent_status'] == 'DisponÃ­vel']['created_time'])

        if(not len(lista_entrada_log) == 0):

            entrada_log = lista_entrada_log[0]

            entrada_log = dt.timedelta(hours = entrada_log.hour, minutes = entrada_log.minute, seconds = entrada_log.second)

            aux_df_horario_primeiro = aux_df_horario.loc[aux_df_horario['Data'] == data]

            entrada_escala = list(aux_df_horario_primeiro['Primeiro HorÃ¡rio'])[0]

            entrada_escala = dt.timedelta(hours = entrada_escala.hour, minutes = entrada_escala.minute, seconds = entrada_escala.second)

            atraso = entrada_log - entrada_escala

            if(atraso >= dt.timedelta(minutes = 15)):
                
                lista_atrasado.append('Atraso')

                #lista_resultado_atraso.append(f"Membro {member} estÃ¡ atrasado por {atraso} minutos (entrou em {data} Ã s {entrada_log} e deveria ter entrado Ã s {entrada_escala}).")

            else: 
                lista_atrasado.append('Dentro do horÃ¡rio')

                #lista_resultado_atraso.append(f"Membro {member} estÃ¡ dentro do horÃ¡rio (entrou em {data} Ã s {entrada_log} e deveria ter entrado Ã s {entrada_escala}).")

        else:
            
            #lista_resultado_atraso.append(f"Membro {member} faltou no dia {data}.")

            lista_atrasado.append('Falta')
```

```python
df_disponibilidade['Primeiro HorÃ¡rio'] = df_horario['Primeiro HorÃ¡rio']
df_disponibilidade['Atraso'] = lista_atrasado

df_disponibilidade
```
