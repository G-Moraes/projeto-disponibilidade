```python
import pandas as pd
import datetime as dt
```

# 1 - ler escala

```python
escala = pd.read_csv('escala.csv')

escala.sort_values(by = ['shift_date', 'member_id'], inplace = True)

escala
```

## 1.1 - tratar dados

```python
start_time_interval = []
end_time_interval = []

for text in escala['time_interval']:
    
    start = text.index('X')
    end1 = text.index('m_', start)
    end2 = text.index('m__', end1)

    startInterval = text[start+1:end1]
    endInterval = text[end1+2: end2]

    startInterval = startInterval.replace('h', ':')
    endInterval = endInterval.replace('h', ':')

    hora_start = dt.datetime.strptime(startInterval, "%H:%M")
    hora_end = dt.datetime.strptime(endInterval, "%H:%M")

    hora_start = dt.timedelta(hours = hora_start.hour, minutes = hora_start.minute)

    hora_end = dt.timedelta(hours = hora_end.hour, minutes = hora_end.minute)

    start_time_interval.append(hora_start)
    end_time_interval.append(hora_end)

    #print(f'começo do intervalo: {hora_start}, fim do intervalo: {hora_end}')

escala.insert(3, 'start_time_interval', start_time_interval)
escala.insert(4, 'end_time_interval', end_time_interval)
```

```python
lista_start_time = []
lista_end_time = []

for start_time, end_time in zip(escala['start_time_interval'], escala['end_time_interval']):

    start_time = str(start_time).split('0 days ')[1]
    end_time = str(end_time).split('0 days ')[1]

    lista_start_time.append(start_time)
    lista_end_time.append(end_time)

escala['start_time_interval'] = lista_start_time

escala['end_time_interval'] = lista_end_time
```

```python
escala.drop(columns = 'time_interval', inplace=True)
```

```python
escala.sort_values(by = ['shift_date', 'member_id', 'start_time_interval', 'end_time_interval'], inplace = True)

escala
```

```python
membros = pd.read_csv('membros.csv')

for index in membros['Data'].index:

    data_dt = dt.datetime.strptime(membros.loc[index]['Data'], "%Y-%m-%d").date()

    membros.loc[index]['Data'] = data_dt

membros.sort_values(by = ['Data', 'member_id'], inplace = True)

membros
```

# 3 - relatorios

```python
relatorio = pd.read_csv('relatorio_log_in_e_out.csv')

relatorio.sort_values(by = ['created_at', 'member_id'], inplace = True)

relatorio
```

## 3.1 - tratar dados

```python
dates = [date.split()[0] for date in relatorio['created_at']]

times = []

for time in relatorio['created_at']:

    time_aux_space = time.split()[1]
    time_aux_dot = time_aux_space.split('.', 1)[0]
    time_aux_plus = time_aux_dot.split('+', 1)
    times.append(time_aux_plus[0])
    
created_date = []
created_time = []

for date, time in zip(dates, times):
    
    short_date = dt.datetime.strptime(date, '%Y-%m-%d').date()
    short_time = dt.datetime.strptime(time, "%H:%M:%S")

    short_time = dt.timedelta(hours = short_time.hour, minutes = short_time.minute, seconds = short_time.second)
    
    created_date.append(short_date)
    created_time.append(short_time)
```

```python
relatorio.insert(2, 'created_date', created_date)
relatorio.insert(3, 'created_time', created_time)
```

```python
relatorio.drop(columns = 'created_at', inplace = True)
```

```python
relatorio.sort_values(by = ['created_date', 'member_id'], inplace = True)

relatorio
```

# 4 - informação A

```python
#celula gera a informação A

#1 - salvo um dataframe com cada membro que estava escalado para atendimento e que dias eles estão
#2 - checo membro um por um, e quantos minutos ele estava escalado para atender em dias especifico
    #2.1 - protip: escalamento é equivalente a meia hora, então só (entrada x 30) para obter minutos
#3 - armazeno essa informação em um dataframe

attending_member = escala.loc[escala['resulting_status'] == 'A']

#unique_attending_members = attending_member['member_id'].unique()

df_atendimento = pd.DataFrame(columns = ['member_id', 'Data', 'escala_atendimento'])

for member in attending_member['member_id'].unique():

    df_attending_member = attending_member.loc[attending_member['member_id'] == member]
    
    for date in df_attending_member['shift_date'].unique():

        concatenado = pd.DataFrame(columns = ['member_id', 'Data', 'escala_atendimento'])

        aux = df_attending_member.loc[df_attending_member['shift_date'] == date]

        concatenado['member_id'] = aux['member_id']
        concatenado['Data'] = aux['shift_date']
        concatenado['escala_atendimento'] = len(aux) * 30

        concatenado.drop_duplicates(inplace = True)

        df_atendimento = pd.concat([df_atendimento, concatenado])
    #break
```

```python
rel_membros_data = {key: [] for key in relatorio['member_id'].unique()}

# 1 - checo o membro, data e status na tabela membros
# 2 - se status == ativo na data, checo o membro na tabela relatorio na data em que ele estava ativo
# 3 - conto o tempo que ele estava online a partir da primeira mudança de status após cada "Disponível"
# 4 - armazeno a informação por dia de cada membro 

#etapa 1, pego todos os membros que estão como ativos
membros_ativos = membros.loc[membros['Status'] == 'Ativo']

#etapa 2

#2.1 - itero sobre os membros ativos
for membro, data in zip(membros_ativos['member_id'], membros_ativos['Data']):

    #2.1.1 - se o membro está na lista de membros ativos
    if membro in rel_membros_data.keys():

        #2.1.2 - salvo todas as datas em que o membro está como ativo
        rel_membros_data[membro].append(data)

#etapa 3
unique_members = relatorio['member_id'].unique()

df_disponibilidade = pd.DataFrame(columns = ['member_id', 'Data', 'Disponibilidade'])

df_disponibilidade['member_id'] = membros['member_id']

df_disponibilidade['Data'] = membros['Data']

#df_disponibilidade['escala_tempo_atendimento'] = 

df_disponibilidade = df_disponibilidade.reset_index(drop = True)

#itero sobre os membros unicos
for member in unique_members:

    #crio um dataframe só com as aparições desse membro unico i.e. suas mudanças de status
    #durante os dias
    relatorio_membro = relatorio.loc[relatorio['member_id'] == member]

    #reseto os indexes
    relatorio_membro = relatorio_membro.reset_index(drop = True)

    #itero sobre o dataframe do membro de uma data em especifico
    for data in rel_membros_data[member]:

        relatorio_membro_data = relatorio_membro.loc[relatorio_membro['created_date'] == data]
        relatorio_membro_data = relatorio_membro_data.reset_index(drop = True)

        #print(relatorio_membro.loc[relatorio_membro['created_date'] == data])

        #itero sobre o dataframe do membro em cada data
        i = 0

        tempo_disponivel_dia = dt.timedelta(0)

        while i < len(relatorio_membro_data):

            #procuro os momentos em que ele estava disponível
            try:

                #checo a primeira ocorrência de Disponível no relatório do membro naquele dia
                index_disponivel = list(relatorio_membro_data['agent_status']).index('Disponível', i, len(relatorio_membro_data))
                
                #print(f'before: index status disponivel: {index_disponivel}, proximo status: {i}, len: {len(relatorio_membro_data)}, {tempo_disponivel_dia}')

                if(index_disponivel == -1 or index_disponivel + 1 >= len(relatorio_membro_data) or i + 1 >= len(relatorio_membro_data)):
                    
                    break
                
                #se o próximo status não ainda estiver no escopo 
                elif(i + 1 <= len(relatorio_membro_data) and index_disponivel + 1 <= len(relatorio_membro_data)):

                    i = index_disponivel + 1

                    tempo_disponivel_dia += relatorio_membro_data.loc[i]['created_time'] - relatorio_membro_data.loc[index_disponivel]['created_time']

                    #print(f'inside if: index status disponivel: {index_disponivel}, proximo status: {i}, len: {len(relatorio_membro_data)}, {tempo_disponivel_dia}')
                          
            except ValueError:
                
                break
        
        aux_tempo = str(tempo_disponivel_dia).split('0 days ')

        if(len(aux_tempo) > 1):
            aux_string = str(tempo_disponivel_dia).split('0 days ')[1]
            aux_dt = dt.datetime.strptime(aux_string, "%H:%M:%S")

        else:
            aux_dt = dt.datetime.strptime(str(tempo_disponivel_dia), "%H:%M:%S")

        tempo_disponivel_dia = dt.timedelta(hours = aux_dt.hour, minutes = aux_dt.minute, seconds = aux_dt.second)
        
        #armazenar a disponibilidade do membro no dia

        disp_index = df_disponibilidade.loc[df_disponibilidade['member_id'] == member]

        disp_index = disp_index.loc[disp_index['Data'] == data].index

        for agente_index in disp_index:

            #print(f"index: {agente_index}, agente: {df_disponibilidade.loc[agente_index]['member_id']}, data: {data}, disponibilidade: {tempo_disponivel_dia}")

            #if(df_disponibilidade.loc[agente_index]['member_id'] == member and df_disponibilidade.loc[agente_index]['Data'] == data):

            df_disponibilidade.loc[agente_index]['Disponibilidade'] = str(int(tempo_disponivel_dia.total_seconds() / 60))
```
